```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi untuk regresi polinomial
def regresi_polinomial(x, y, derajat):
    # Menghitung koefisien polinomial
    koefisien = np.polyfit(x, y, derajat)
    # Membuat fungsi polinomial dari koefisien
    polinomial = np.poly1d(koefisien)
    return polinomial

# Data contoh 
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([1, 3, 5, 7, 9, 11])

# Menggunakan regresi polinomial dengan derajat 2
derajat = 2
polinom = regresi_polinomial(x, y, derajat)

# Menghitung nilai y hasil regresi
y_polinom = polinom(x)

# Plot data asli dan hasil regresi
plt.scatter(x, y, color='red', label='Data Asli')
plt.plot(x, y_polinom, label=f'Regresi Polinomial Derajat {derajat}', color='blue')

# Menambahkan label dan judul
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regresi Polinomial Sederhana')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi untuk regresi eksponensial
def regresi_eksponensial(x, y):
    # Mengambil logaritma alami dari y
    log_y = np.log(y)
    # Menghitung koefisien linear dari logaritma y
    koefisien = np.polyfit(x, log_y, 1)
    # Mendapatkan nilai a dan b
    a = np.exp(koefisien[1])
    b = koefisien[0]
    # Membuat fungsi eksponensial
    def eksponensial(t):
        return a * np.exp(b * t)
    return eksponensial

# Data contoh 
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([1, 3, 5, 7, 9, 11])

# Menggunakan regresi eksponensial
eksponensial = regresi_eksponensial(x, y)

# Menghitung nilai y hasil regresi
y_eksponensial = eksponensial(x)

# Plot data asli dan hasil regresi
plt.scatter(x, y, color='red', label='Data Asli')
plt.plot(x, y_eksponensial, label='Regresi Eksponensial', color='blue')

# Menambahkan label dan judul
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regresi Eksponensial Sederhana')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Fungsi untuk model logistik
def model_logistik(x, L, k, x0):
    return L / (1 + np.exp(-k * (x - x0)))

# Fungsi untuk regresi logistik
def regresi_logistik(x, y):
    # Menyediakan tebakan awal untuk parameter L, k, dan x0
    initial_guess = [max(y), 1, np.median(x)]
    # Menggunakan curve_fit untuk menyesuaikan model logistik dengan data
    popt, _ = curve_fit(model_logistik, x, y, p0=initial_guess)
    # Mendapatkan parameter yang dioptimalkan
    L, k, x0 = popt
    # Mengembalikan fungsi logistik dengan parameter yang dioptimalkan
    def logistik(t):
        return model_logistik(t, L, k, x0)
    return logistik

# Data contoh 
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([1, 3, 5, 7, 9, 11])

# Menggunakan regresi logistik
logistik = regresi_logistik(x, y)

# Menghitung nilai y hasil regresi
y_logistik = logistik(x)

# Plot data asli dan hasil regresi
plt.scatter(x, y, color='red', label='Data Asli')
plt.plot(x, y_logistik, label='Regresi Logistik', color='blue')

# Menambahkan label dan judul
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regresi Logistik Sederhana')
plt.legend()
plt.show()
```
