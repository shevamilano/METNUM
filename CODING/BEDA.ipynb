```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi yang akan kita hitung beda majunya
def fungsi(x):
    return np.sin(x)  # Misalnya, f(x) = sin(x)

# Beda maju untuk fungsi f pada titik x
def beda_maju(f, x, h):
    return (f(x + h) - f(x)) / h

# Rentang x dan nilai h
x = np.linspace(0, 2 * np.pi, 100)  # 100 titik dari 0 hingga 2π
h = 0.01  # Nilai h kecil untuk mendekati turunan sebenarnya

# Hitung nilai fungsi f(x) dan beda maju f'(x)
f_x = fungsi(x)
f_prim_x = beda_maju(fungsi, x, h)

# Plot fungsi f(x) dan beda maju f'(x)
plt.figure(figsize=(10, 6))
plt.plot(x, f_x, label='f(x) = sin(x)')
plt.plot(x, f_prim_x, label="f'(x) beda maju", linestyle='--')

# Tambahkan judul dan label
plt.title('Beda Maju dan Fungsi Aslinya')
plt.xlabel('x')
plt.ylabel('Nilai')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi yang akan kita hitung beda mundurnya
def fungsi(x):
    return np.sin(x)  # Misalnya, f(x) = sin(x)

# Beda mundur untuk fungsi f pada titik x
def beda_mundur(f, x, h):
    return (f(x) - f(x - h)) / h

# Rentang x dan nilai h
x = np.linspace(0, 2 * np.pi, 100)  # 100 titik dari 0 hingga 2π
h = 0.01  # Nilai h kecil untuk mendekati turunan sebenarnya

# Hitung nilai fungsi f(x) dan beda mundur f'(x)
f_x = fungsi(x)
f_prim_x = beda_mundur(fungsi, x, h)

# Plot fungsi f(x) dan beda mundur f'(x)
plt.figure(figsize=(10, 6))
plt.plot(x, f_x, label='f(x) = sin(x)')
plt.plot(x, f_prim_x, label="f'(x) beda mundur", linestyle='--')

# Tambahkan judul dan label
plt.title('Beda Mundur dan Fungsi Aslinya')
plt.xlabel('x')
plt.ylabel('Nilai')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi yang akan kita hitung beda pusatnya
def fungsi(x):
    return np.sin(x)  # Misalnya, f(x) = sin(x)

# Beda pusat untuk fungsi f pada titik x
def beda_pusat(f, x, h):
    return (f(x + h) - f(x - h)) / (2 * h)

# Rentang x dan nilai h
x = np.linspace(0, 2 * np.pi, 100)  # 100 titik dari 0 hingga 2π
h = 0.01  # Nilai h kecil untuk mendekati turunan sebenarnya

# Hitung nilai fungsi f(x) dan beda pusat f'(x)
f_x = fungsi(x)
f_prim_x = beda_pusat(fungsi, x, h)

# Plot fungsi f(x) dan beda pusat f'(x)
plt.figure(figsize=(10, 6))
plt.plot(x, f_x, label='f(x) = sin(x)')
plt.plot(x, f_prim_x, label="f'(x) beda pusat", linestyle='--')

# Tambahkan judul dan label
plt.title('Beda Pusat dan Fungsi Aslinya')
plt.xlabel('x')
plt.ylabel('Nilai')
plt.legend()
plt.grid(True)
plt.show()
```
